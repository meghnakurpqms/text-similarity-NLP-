```python
import pandas as pd 
class CorpusReader:
    def read_data(self,path):
        data = pd.read_csv(path, sep="\t", header=None,error_bad_lines=False)
        return data
        
```

```python
data = CorpusReader()
```

```python
dev = 'dev-set.txt'
train = 'train-set.txt'
dev_set = data.read_data(dev)
train_set = data.read_data(train)
```
