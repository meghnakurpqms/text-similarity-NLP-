```python
import nltk
lemmatizer = nltk.stem.WordNetLemmatizer()
```

```python
def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in text]
```

```python
dev_set2[6] = dev_set2[4].apply(lemmatize_text)
dev_set2[7] = dev_set2[5].apply(lemmatize_text)
```
