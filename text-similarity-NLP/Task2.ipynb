```python
import spacy
```

```python
nlp = spacy.load("en")
```

```python
def dependency_parsing_tree(text):
    res = []
    sent = nlp(text)
    for token in sent:
        res.extend([token.text, token.tag_, token.head.text, token.dep_])
    return res
        
```

```python
dev_set4[10] = dev_set4[1].apply(dependency_parsing_tree)
dev_set4[11] = dev_set4[2].apply(dependency_parsing_tree)
```

```python
import collections 
def syn_hyper(text):
    syn = []
    for i in text:
        d = {}
        for ss in wn.synsets(i):
            for hyper in ss.hypernyms():
#             syn.extend([ss, ss.hypernyms()])
                d[i] = [ss,hyper]
        syn.append(d)
    return syn
```

```python
import collections 
def syn_hypo(text):
    syn = []
    for i in text:
        d = {}
        for ss in wn.synsets(i):
            for hypo in ss.hyponyms():
#             syn.extend([ss, ss.hypernyms()])
                d[i] = [ss,hypo]
        syn.append(d)
    return syn
```
