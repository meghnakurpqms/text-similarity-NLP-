```python
from nltk.corpus import stopwords

def cosine_similarity_hypo(f):
    X = set(list(f[1]))
    Y = set(list(f[2]))

    sw = stopwords.words('english')
    
    X_set = {w for w in X if not w in sw}  
    Y_set = {w for w in Y if not w in sw}
    
    rvector = X.union(Y)

    l1 = []
    l2 = []

    for w in rvector: 
        if w in X: l1.append(1) # create a vector 
        else: l1.append(0) 
        if w in Y: l2.append(1) 
        else: l2.append(0) 
    c = 0

    # cosine formula  
    for i in range(len(rvector)): 
            c+= l1[i]*l2[i]
    
    if sum(l1)*sum(l2) == 0:
        cosine = 0
    else:
        cosine = c / float((sum(l1)*sum(l2))**0.5) 
    return cosine 
```

```python
from nltk.corpus import stopwords

def cosine_similarity_hyper(f):
    X = set(list(f[6]))
    Y = set(list(f[7]))
    
    sw = stopwords.words('english')
    
    X_set = {w for w in X if not w in sw}  
    Y_set = {w for w in Y if not w in sw}
    
    
    rvector = X.union(Y)

    l1 = []
    l2 = []

    for w in rvector: 
        if w in X: l1.append(1) # create a vector 
        else: l1.append(0) 
        if w in Y: l2.append(1) 
        else: l2.append(0) 
    c = 0

    # cosine formula  
    for i in range(len(rvector)): 
            c+= l1[i]*l2[i]
    
    if sum(l1)*sum(l2) == 0:
        cosine = 0
    else:
        cosine = c / float((sum(l1)*sum(l2))**0.5) 
    return cosine 
```

```python
def jaccard_similarity(f):
    
    list1 = list(f[6])
    list2 = list(f[7])
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection) / union
```
