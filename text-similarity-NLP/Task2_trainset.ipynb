# Lemmatize

```python
import nltk
lemmatizer=nltk.stem.WordNetLemmatizer()
```

```python
#lemmatize every word in the text
def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in text]
```

# Dependency Parse tree

```python
import spacy
```

```python
# load in english
nlp = spacy.load("en")
```

```python
nlp("He saw a red rose.")
```

```python
def dependency_parse_tree(text):
    res = []
    sent = nlp(text)
    for token in sent:
        res.extend([token.text,token.tag_,token.head.text,token.dep_])
    return res
```

```python
TrainSet[10] = TrainSet[1].apply(dependency_parse_tree)
TrainSet[11] = TrainSet[2].apply(dependency_parse_tree)
```

# Wordnet

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.corpus import wordnet as wn 
```

```python
#Setting synsets and hypernms
import collections
def synHyper_set(text):
    syn = []
    for i in text:
        d = {}
        for ss in wn.synsets(i):
            for hyper in ss.hypernyms():
                d[i] = [ss, hyper]
        syn.append(d)
    return syn
```

```python
TrainSet[12] = TrainSet[6].apply(synHyper_set)
```

```python
TrainSet
```

```python
#Setting synsets and hyponym
import collections
def synHypo_set(text):
    syn = []
    for i in text:
        d = {}
        for ss in wn.synsets(i):
            for hypo in ss.hyponyms():
                d[i] = [ss,hypo]
        syn.append(d)
    return syn
```

```python
TrainSet[13] = TrainSet[6].apply(synHypo_set)
#dev_set4[14] = dev_set4[6].apply(syn_hypo)
```
